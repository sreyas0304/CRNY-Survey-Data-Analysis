```python
# Define a function to count non-null values in the specified columns
def count_non_null(row):
    caregiving_columns = ['p46_caregiving1', 'p46_caregiving2', 'p46_caregiving3']
    if row['caregiving_status'] == 'Yes':
        return row[caregiving_columns].notnull().sum()
    else:
        return 0

# Apply the function row-wise to create the new column caregiving_count
new_pdata['caregiving_count'] = new_pdata.apply(count_non_null, axis=1)

# Display the DataFrame with the new column
print(new_pdata[['p46_caregiving1', 'p46_caregiving2', 'p46_caregiving3', 'caregiving_status', 'caregiving_count' ]])
```
